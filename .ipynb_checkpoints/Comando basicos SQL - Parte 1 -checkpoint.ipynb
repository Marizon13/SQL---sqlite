```python
# Criando a conexão
import sqlite3
con = sqlite3.connect('BaseDados.db')
```

```python
# Criando um cursor
cur = con.cursor()
```

```python
# Buscando todas as informações da tabela
cur.execute('SELECT * FROM dados')
resultado = cur.fetchall()
```

```python
# Salvando em um DataFrame do pandas
import pandas as pd

resultado = pd.DataFrame(resultado)
resultado.head()
```

### Selecionando dados



```python
# Selecionando todos os dados da tabela


cur.execute('SELECT nome_aluno,cod_matricula FROM dados') # usando o filtro nessa linha de código
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()


cur.execute('SELECT * FROM dados')
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = colunas
resultado.head()
```

```python
# selecione apenas as colunas nome_aluno e cod_matricula

cur.execute('SELECT nome_aluno,cod_matricula FROM dados') # usando o filtro nessa linha de código
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()
```

```python
# selecionando apenas as colunas nome_aluno e e-mail

cur.execute('SELECT nome_aluno,[e-mail] as email FROM dados') # usando o filtro nessa linha de código # caso dê erro por conta do e- é só passar em [] e atribui com as
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()
```

```python
comando = 'SELECT nome_aluno,[e-mail] as email FROM dados' # atribuindo 1 variavel na formula do select
cur.execute(comando)
resultado = cur.fetchall()
resultado = pd.DataFrame(resultado)
resultado.columns = [i[0] for i in cur.description]
resultado.head()
```

```python
# criando função  para facilitar e não ficar escrevendo toda a linha de código novamente
```

```python
def executa_sql (comando):
    cur.execute(comando)
    resultado = cur.fetchall()
    resultado = pd.DataFrame(resultado)
    resultado.columns = [i[0] for i in cur.description]
    print(resultado.shape)
    display(resultado.head())
    return resultado
```

```python
resultado = executa_sql('SELECT nome_aluno,[e-mail] as email FROM dados')
```
